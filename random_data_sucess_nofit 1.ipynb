```python
import random
import matplotlib.pyplot as plt


distance_from_earth_range = (4.24, 100000)  # From Proxima Centauri(outside of our solar system) to the other side of the Milky Way
esi_range = (0.5, 1.0)  # Adjusted range for Earth Similarity Index, minimum 0.5

# Number of random data points you want to generate
num_data_points = 258

# Generate random test data
random_test_data = []
for _ in range(num_data_points):
    # Generate random distance and ESI values within the defined ranges
    distance = random.uniform(*distance_range)
    esi = random.uniform(*esi_range)
    
    # Append the generated data as a tuple (distance, ESI) to the list
    random_test_data.append((distance, esi))

# Unzip the random test data into separate lists for distances and ESI values
distances, esis = zip(*random_test_data)

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(distances, esis, color='blue', alpha=0.5)
plt.xlabel('Distance (light-years) from Earth')
plt.ylabel('Earth Similarity Index (ESI)')
plt.title('Randomly Generated Test Data')
plt.grid(True)
plt.show()
```
